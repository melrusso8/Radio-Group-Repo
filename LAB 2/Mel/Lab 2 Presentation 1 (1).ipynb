## Calculate Average Power Spectra of s_cold 

```python
# COLD DATA
freq = np.fft.fftfreq(cold_data.shape[1], 1/62.5e6)
freq = np.fft.fftshift(freq)/1e6

blocks = 8000
avg_power_array = []
for n in range (0, blocks):
    data_new = cold_data[n]
    ft = np.fft.fft(data_new)
    ft = np.fft.fftshift(ft)
    power = abs(ft)**2
    avg_power_array.append(power)
    
avg_power_array = np.mean(avg_power_array, axis = 0)
s_off = avg_power_array
```

## Plot Average Power Spectra of s_cold

```python
plt.figure(figsize=(10,5))
plt.plot(freq + 1230 + 190, avg_power_array, "mediumpurple", linewidth=1)
plt.yscale("log")
plt.xlabel("Frequency (MHz)", fontsize=15)
plt.ylabel("Power", fontsize=15)
plt.title("Average Power Spectrum of Cold Data", fontsize=25)
plt.legend(fontsize=15)

#plt.xlim(-2, 2)
#plt.vlines(1420.4058, ymin=10e8, ymax=10e11, linewidth=.5)
```

# Calculate Gain

### find T_sys

```python
# temperature
T_rx = 300 #kelvin
T_sky = 10 #between 10-100K
T_sys = T_rx - T_sky
T_sys
```

### filter out spikes 

```python
# shape
g_freq = freq + 1230 + 190
s_line = s_on / s_off #probably wrong. ratio of on observation (hydrogen there), over off observation (no H)

filtered_s_on = np.array([s for s in s_on if not (g_freq[list(s_on).index(s)]<1390 or g_freq[list(s_on).index(s)]>1451 or (g_freq[list(s_on).index(s)]< 1421 and g_freq[list(s_on).index(s)]>1419))])
filtered_s_off = np.array([s for s in s_off if not (g_freq[list(s_off).index(s)]<1390 or g_freq[list(s_off).index(s)]>1451 or (g_freq[list(s_off).index(s)]< 1421 and g_freq[list(s_off).index(s)]>1419))])
print(len(filtered_s_on), len(filtered_s_off))                     
```

```python
filtered_s_off
```

### calculate G and T_line

```python
G = T_sys/(np.sum(filtered_s_off-filtered_s_on)) * np.sum(filtered_s_on)
T_line = s_line * G #power spectra in temperature units 
```
